### Grouping the data on the basis of classification variable

```python
df.groupby(['Region']).count()
```

```python
df.groupby(['Position']).count()
```

### Summarizing data by classification variable

```python
print("Mode Of Region : "+str(stats.mode(df['Region'])))
print("Mode of Position : "+str(stats.mode(df['Position'])))
print("Mean of Age : "+str(np.mean(df['Age'])))
print("Median of Age : "+str(np.median(df['Age'])))
print("Mode of Age : "+str(stats.mode(df['Age'])))
```

### Reshaping data into wide/long format

```python
df.pivot(index='Rg No.', columns='Region', values='Name')
```

```python
df.pivot(index='Rg No.', columns='Position', values='Name')
```
