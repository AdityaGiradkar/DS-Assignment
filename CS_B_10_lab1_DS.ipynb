```
first_name = data.loc[:, 'First Name']
last_name = data.loc[:, 'Last Name']

df1 = pd.DataFrame({"First Name": first_name})

display(df1.sample(n = 5))

df2 = pd.DataFrame({"Last Name" : last_name})

display(df2.sample(n = 5))

res = pd.concat([df1, df2], ignore_index=True)

display(res.sample(n = 5))
```

```
City = data.loc[:,'City']
State = data.loc[:,'State']

left = pd.DataFrame({
   'id':range(5000),
   'City': City})

right = pd.DataFrame({
   'id':range(5000),
   'State': State})
# print(right)

result = pd.merge(left, right, on="id")
display(result)
```
