### Importing Libraries

```python
import pandas as pd
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt 
```

### Importing the dataset

```python
df = pd.read_csv("dataset/adult.csv")
df
```

### Replacing "?" with NaN

```python
df = df.replace('?',np.nan)
df
```

### Checking the Missing data

```python
df.isna()
```

### Counting the Missing Values

```python
df.isna().sum()
```

### Droping the Missing Data

```python
dropdf=df.dropna()
dropdf.isna().sum()
```

```python
print(df.shape)
print(dropdf.shape)
```

### Fill in Missing values with Mode

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputer.fit(df.iloc[:,:])
df.iloc[:,:] = imputer.transform(df.iloc[:,:])
df
```

```python
df.isna().sum()
```

## Transform the non-numeric columns into numeric
### Lable Encoding

```python
df = df.rename(columns = {'marital.status': 'maritalstatus', 'native.country': 'nativecountry'}, inplace = False)
from sklearn.preprocessing import LabelEncoder
l1 = LabelEncoder()
l1.fit(df['workclass'])
df.workclass = l1.transform(df.workclass)
l1.fit(df['education'])
df.education = l1.transform(df.education)
l1.fit(df['maritalstatus'])
df.maritalstatus = l1.transform(df.maritalstatus)
l1.fit(df['occupation'])
df.occupation = l1.transform(df.occupation)
l1.fit(df['relationship'])
df.relationship = l1.transform(df.relationship)
l1.fit(df['race'])
df.race = l1.transform(df.race)
l1.fit(df['sex'])
df.sex = l1.transform(df.sex)
l1.fit(df['nativecountry'])
df.nativecountry = l1.transform(df.nativecountry)
df
```
