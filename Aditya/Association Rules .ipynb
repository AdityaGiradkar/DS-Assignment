```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```python
df = df.sample(5000)
df.head()
```

```python
#Data Preprocessing
df['Description'] = df['Description'].str.strip()
df.dropna(axis = 0, subset = ['InvoiceNo'], inplace=True)
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
df = df[~df['InvoiceNo'].str.contains('C')]
df
```

```python
df1 = df.groupby(['InvoiceNo','InvoiceDate'])
df1.head()
```

```python
df1.first()
```

```python
transactions = []
for name,group in df1:
    transactions.append([str(group.values[j,2]) for j in range(group.shape[0])])
```

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.001, min_confidence = 0.8, min_lift =
4,min_length=2,max_length=3)
```

```python
results = list(rules)
print("{} rules are derived.".format(len(results)))
```

```python
for i in range(0, 10):
  print(results[i][0])
```

```python
for i in range(10) :
  ab = results[i][0]
  items = [x for x in ab]
  print(items[0]+' -> '+items[1])
  print("Support: " + str(results[i][1]))
  print("Confidence: " + str(results[i][2][0][2]))
  print("Lift: " + str(results[i][2][0][3]))
  print('------------------------------------')
```

```python
#Convert to DataFrame
def check(results):
  lhs = [tuple(i[2][0][0])[0] for i in results]
  rhs = [tuple(i[2][0][1])[0] for i in results]
  s = [i[1] for i in results]
  c = [i[2][0][2] for i in results]
  l = [i[2][0][3] for i in results]
  return list(zip(lhs, rhs, s, c, l))
  
resultsDF = pd.DataFrame(check(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support','Confidence', 'Lift'])
```
