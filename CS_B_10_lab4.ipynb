### Group data in the dataframe by InvoiceNo and Invoice Date

```python
df1 = df.groupby(['InvoiceNo','InvoiceDate']).groups
print(df1)
```

### Total number of Transactions 

```python
df1 = df.groupby(['InvoiceNo','InvoiceDate'])
df1.head()
```

```python
des = df['Description']
transactions = []
for x, y in df1.groups.items():
    y = list(y)
    transactions.append([des[y[j]] for j in range(len(y))])
transactions[0]
```

### Total Number of Transactions with 1 item, 2 items , 3 items and 8 items

```python
i1 = i2 = i3 = i8 = 0
for i in transactions:
    if len(i)==1:
        i1 = i1 + 1
    if len(i)==2:
        i2 = i2 + 1
    if len(i)==3:
        i3 = i3 + 1
    if len(i)==8:
        i8 = i8 + 1
print("No. transaction containing 1 item : "+str(i1))
print("No. transaction containing 2 item : "+str(i2))
print("No. transaction containing 3 item : "+str(i3))
print("No. transaction containing 8 item : "+str(i8))
```

```python
trans = []
# print(transactions)
for i in range(10):
    trans.append(transactions[i])
trans[1]
```

### Training the Apriori model on the dataset

```python
from apyori import apriori
rules = apriori(trans, min_support = 0.001, min_confidence = 0.8,max_length = 3)
results = list(rules)
print("There are {} RULES derived.".format(len(results)))
for i in range(5):
  print(results[i])
```

###Display rules LHS RHS Support Confidence Lift

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
resultsinDataFrame
```

###Print only top 10 rules wrt lift in following format

```python
# A -> B [support=s, confidence=c, lift=l]

i=0
for item in results:
  if i==10:
    break
  pair=item[0]
  items=[x for x in pair]
  print(items[0] + '->' + items[1] + ' [Support=' + str(item[1]) + ', confidence=' + str(item[2][0][2]) + ', lift=' + str(item[2][0][3]) + ']')
  i=i+1
```
