# Finding mean, median, mode and midrange

```python
print("Sepal Length in cm")
print("\tMean = "+str(np.mean(df['sepal_length'])))
print("\tMedian = "+str(np.median(df['sepal_length'])))
print("\tMode = "+str(stats.mode(df['sepal_length'])))
print("\tMidRange = ",(min(df['sepal_length'])+max(df['sepal_length']))/2)
print("\n")

print("Sepal Width in cm")
print("\tMean = "+str(np.mean(df['sepal_width'])))
print("\tMedian = "+str(np.median(df['sepal_width'])))
print("\tMode = "+str(stats.mode(df['sepal_width'])))
print("\tMidRange = ",(min(df['sepal_width'])+max(df['sepal_width']))/2)
print("\n")


print("Petal Length in cm")
print("\tMean = "+str(np.mean(df['petal_length'])))
print("\tMedian = "+str(np.median(df['petal_length'])))
print("\tMode = "+str(stats.mode(df['petal_length'])))
print("\tMidRange = ",(min(df['petal_length'])+max(df['petal_length']))/2)
print("\n")


print("Petal Width in cm")
print("\tMean = "+str(np.mean(df['petal_width'])))
print("\tMedian = "+str(np.median(df['petal_width'])))
print("\tMode = "+str(stats.mode(df['petal_width'])))
print("\tMidRange = ",(min(df['petal_width'])+max(df['petal_width']))/2)
print("\n")
```

# Removing the outliers using IQR

```python
Q1 =  df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
gk=df[(df< Q1-1.5*IQR ) | (df > Q3+1.5*IQR)]
ot = gk.notnull()
```

# Box plot with the dataset after the removal of outliers

```python
plt.rcParams["figure.figsize"] = (10,8)
otRI = list(ot[ot['RI']==True].index)
df1 = df.drop(otRI)

otNa = list(ot[ot['Na']==True].index)
df2 = df.drop(otNa)

otMg = list(ot[ot['Mg']==True].index)
df3 = df.drop(otMg)

plt.subplots_adjust(hspace=0.5, wspace=1)
plt.subplot(1,3,1)
plt.boxplot(df1['RI'],patch_artist=True)
plt.title('RI')
plt.subplot(1,3,2)
plt.boxplot(df2['Na'],patch_artist=True)
plt.title('Na')
plt.subplot(1,3,3)
plt.boxplot(df3['Mg'],patch_artist=True)
plt.title('Mg')
plt.show()


otAl = list(ot[ot['Al']==True].index)
df4 = df.drop(otAl)

otSi = list(ot[ot['Si']==True].index)
df5 = df.drop(otSi)

otK = list(ot[ot['K']==True].index)
df6 = df.drop(otK)


plt.subplots_adjust(hspace=0.5, wspace=1)
plt.subplot(1,3,1)
plt.boxplot(df4['Al'],patch_artist=True)
plt.title('Al')
plt.subplot(1,3,2)
plt.boxplot(df5['Si'],patch_artist=True)
plt.title('Si')
plt.subplot(1,3,3)
plt.boxplot(df6['K'],patch_artist=True)
plt.title('K')
plt.show()

otCa = list(ot[ot['Ca']==True].index)
df7 = df.drop(otCa)

otBa = list(ot[ot['Ba']==True].index)
df8 = df.drop(otBa)

otFe = list(ot[ot['Fe']==True].index)
df9 = df.drop(otFe)

plt.subplots_adjust(hspace=0.5, wspace=1)
plt.subplot(1,3,1)
plt.boxplot(df7['Ca'],patch_artist=True)
plt.title('Ca')
plt.subplot(1,3,2)
plt.boxplot(df8['Ba'],patch_artist=True)
plt.title('Ba')
plt.subplot(1,3,3)
plt.boxplot(df9['Fe'],patch_artist=True)
plt.title('Fe')
plt.show()
```
